```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(21)
print(x)
```

```python
x = np.arange(0,21)
print(x)
```

```python
y = 2 * x + 3
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line with Slope 2 and y-intercept 3")
plt.show()
```

```python
plt.plot(x,y, linestyle = "dashed")
plt.title("Line with Slope 2 and y-intercept 3")
plt.grid()
plt.show()
```

```python
plt.plot(x,y,
         linestyle = "dashdot",
         color = "crimson")
plt.title("Line with Slope 2 and y-intercept 3")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, linestyle = "dashdot", color = "crimson", linewidth = 3)
plt.title("Line with Slope 2 and y-intercept 3")
plt.grid()
plt.show()
```

```python
plt.style.use('classic')
plt.plot(x,y, linestyle = "dashdot", color = "crimson", linewidth = 3)
plt.title("Line with Slope 2 and y-intercept 3")
plt.grid()
plt.show()
```

```python
from matplotlib import style
  
print(plt.style.available)
```

```python
small_errors = np.random.normal(0,1,21)
large_errors = np.random.normal(0,5,21)
huge_errors = np.random.normal(0,15,21)

y_small_errors = y + small_errors
y_large_errors = y + large_errors
y_huge_errors = y + huge_errors
```

```python
plt.style.use('fivethirtyeight')
plt.plot(x,y,
         linestyle = "dashed",
         color = "black",
         linewidth = 2)
plt.scatter(x,y_small_errors, color = "red", label = "Small Errors")
plt.scatter(x,y_large_errors, color = "green", label = "Large Errors")
plt.scatter(x,y_huge_errors, color = "blue", label = "Huge Errors")
plt.legend()
plt.title("Line with Slope 2 and y-intercept 3")

plt.show()
```

```python
y2 = 2.5 * x + 1
plt.plot(x,y,color = "red", label = "2x + 3")
plt.plot(x,y2,color = "blue", label = "2.5x + 1")
plt.legend()
plt.title("Two Lines")
plt.show()
```
